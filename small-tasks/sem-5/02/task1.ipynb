```python
import numpy as np
import math
import numpy
import matplotlib.pyplot as plt
```

```python
delta_u_target = 10**(-3)
t_target = 0.5
def u_even (t_rad):
    return np.sin(t_rad)
def u_non_even(t_rad):
    return np.cos(t_rad)
```

```python
def calc (t, n):
    res = 0
    for i in range (n):
        pow = t**i
        sign = (-1)**((i//2) % 2)
        val = u_non_even(0) if (i % 2) else u_even(0)
        factorial = math.factorial(i)
        res += pow * sign * val / factorial
    return res
def abs_error(t, n):
    return abs(u_even(t) - calc(t, n))
```

```python
for i in range (15):
    err = abs_error(t_target, i)
    if (err <= delta_u_target):
        print(f"First suitable N: {i}. Error: {err}")
        break
```

```python
max_n = 15
errors = []
for i in range (0, max_n):
    errors.append(abs_error(t_target, i))
ns = np.arange(max_n)

plt.figure(figsize=(16,9), dpi=90)
plt.title("Plot Error(N)")
plt.xlabel("N")
plt.ylabel("Error")
plt.plot(ns, errors)
plt.show()
```
