```python
import numpy as np
import math
import matplotlib.pyplot as plt

def f1(X):
    return math.sqrt(1 - X**2)
def f2(X):
    return math.tan(X)

plt.figure(figsize = (5, 5))

X = np.arange(-1, 1, 0.001)
Y = [f1(i) for i in X]
plt.plot(X, Y, 'green')

Y = [-i for i in Y]
plt.plot(X, Y, 'green')

Y = [f2(i) for i in X]
plt.plot(X, Y, 'red')

SolutionX = [0.65, -0.65]
SolutionY = [0.76, -0.76]
plt.scatter(SolutionX, SolutionY, marker = "X")

plt.grid()
plt.show()
```

solution:

$x = \pm arctg \sqrt{1 - x^2} = f(x)$

Iteration process for positive solution

$x_{i + 1} = arctg \sqrt{1 - x_i^2}$



```python
def getNextX(X):
    return math.atan(math.sqrt((1 - X**2)))

def getSolution(X0, Epsilon):
    Xk = X0
    Xk1 = getNextX(Xk)
    while (math.fabs(Xk1 - Xk) > Epsilon):
        Xk = Xk1
        Xk1 = getNextX(Xk)
    return Xk1

Epsilon = 1e-6
X0 = 0.6
X = getSolution(X0, Epsilon)
print("Solution: x = +-", format(X, '.3f'), ", y = +-", format(math.tan(X),  '.3f'))
```

## Task 2

Для нелинейной системы уравнений

$xy - x^2 = 1,03$     - *Red line*

$−2x^3 + y^2 = 1,98$

```python
def f3(X):
    return (1.03 + X**2) / X
def f4(X):
    return math.sqrt(1.98 + 2 * X**3)

plt.figure(figsize = (5, 5))

X = np.arange(-0.99, 2, 0.001)
Y = [f4(i) for i in X]
plt.plot(X, Y, 'green')

Y = [-i for i in Y]
plt.plot(X, Y, 'green')

X1 = np.arange(-1.5, -0.1, 0.001)
X2 = np.arange(0.1, 2, 0.001)
Y = [f3(i) for i in X1]
plt.plot(X1, Y, 'red')
Y = [f3(i) for i in X2]
plt.plot(X2, Y, 'red')

plt.grid()
plt.show()
```

There is a solution on the interval:

x: $[0.8, 1.2]$

y: $[1.8, 2.2]$


```python
def getNextXY(X, Y):
    X1 = ((Y**2 - 1.98) / 2)**(1.0 / 3)
    Y1 = X + 1.03 / X
    return X1, Y1

def getSolution(X0, Y0, Epsilon):
    Xk, Yk   = X0, Y0
    Xk1, Yk1 = getNextXY(Xk, Yk)
    Count = 0
    while (math.sqrt((Xk1 - Xk)**2 + (Yk1 - Yk)**2) > Epsilon):
        Count += 1
        Xk, Yk   = Xk1, Yk1
        Xk1, Yk1 = getNextXY(Xk, Yk)
    return Xk1, Yk1, Count

Epsilon = 1e-4
X0, Y0 = 1.0, 2.0
X, Y, Count = getSolution(X0, Y0, Epsilon)
print("Iteration Count:", Count)
print("Solution: x =", format(X, '.3f'), ", y =", format(Y,  '.3f'))
```
